El perímetro de un triángulo isósceles es de $10 cm$. ¿Cuánto deben medir sus lados para que el volumen del cuerpo generado por la rotación del triángulo en torno a su base sea el mayor posible? (Volumen de un cono $= \frac{1}{3}\pi r^2 h$, donde $r$ es el radio de la base y $h$ la altura del cono).

1. Definir criterio de optimización
   - Maximizar
2. Variables que afectan la función de costo
   - $x$, $y$, $h$
3. Función de costo
   - $= \frac{1}{3}\pi r^2 h$
4. Información adicional
   - Perímetro triángulo = 10 cm 
   - Volumen = $= \frac{1}{3}\pi r^2 h$



```python
import sympy as sp
import numpy as np

sp.var("x")
sp.var("y")
sp.var("h")

perimetro = 2*x + y - 10
perimetro
```

```python
xp_sol = sp.solve(perimetro, x)[0]
xp_sol
```

```python
radio = y/2
radio
```

```python
altura = sp.sqrt(x**2 - (y/2)**2)
altura
```

```python
val_h = altura.subs(x, xp_sol)
val_h
```

# Volumen = $= \frac{1}{3}\pi r^2 h$

```python
volumen = (1/3)*sp.pi*(radio**2)*h
volumen
```

```python
vol = sp.diff(volumen, y)
vol
```

```python
dvol = sp.solve(vol, y)[1]
dvol
```

```python
val_x = xp_sol.subs(y, dvol)
val_x
```

Disponemos de una barra de aluminio de 6 metros para construir una portería de fútbol. Si queremos que el área de la portería sea máxima, ¿cuánto deben medir los postes y el travesaño?

1. Definir criterio de optimización
   - Maximizar
2. Variables que afectan la función de costo
   - $x$, $y$
3. Función de costo
   - $f(x) = 2x + y$
4. Información adicional
   - Barra de aluminio = 6m
   

```python
import sympy as sp
import numpy as np

sp.var("x")
sp.var("y")

perimetro = 2*x + y - 6
perimetro
```

```python
x_sol = sp.solve(perimetro, x)[0]
x_sol
```

```python
area = x*y
area_y = area.subs(x, x_sol)
area_y
```

```python
darea_y = sp.diff(area_y, y)
darea_y
```

```python
df = sp.solve(darea_y, y)[0]
df
```

```python
val_x = x_sol.subs(y, df)
val_x
```

```python
2*val_x + df
```

### Para maximizar el área, se necesita que los lados de la porteria midan $3/2 m$ y que el traveseño mida $3 m$.
